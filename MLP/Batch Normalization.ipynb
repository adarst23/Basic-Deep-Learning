```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=200, random_state=1, noise=0.3)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow', edgecolors='black')
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense, BatchNormalization
from tensorflow.keras import Sequential
```

```python
model1 = Sequential()

model1.add(Dense(2, activation='relu', input_dim=2))
model1.add(Dense(2, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history1 = model1.fit(X, y, epochs=200, validation_split = 0.2)
```

```python
model2 = Sequential()

model2.add(Dense(2, activation='relu', input_dim=2))
model2.add(BatchNormalization())
model2.add(Dense(2, activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(1, activation='sigmoid'))

model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history2 = model2.fit(X, y, epochs=200, validation_split = 0.2)
```

```python
plt.plot(history1.history['accuracy'], color='black')
plt.plot(history2.history['accuracy'], color='green')
```
